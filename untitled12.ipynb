```python
from sklearn.preprocessing import MultiLabelBinarizer

data["skills_list"] = data["required_skills"].apply(lambda x: x.split(","))

mlb = MultiLabelBinarizer()
skills_encoded = pd.DataFrame(
    mlb.fit_transform(data["skills_list"]),
    columns=mlb.classes_,
    index=data.index
)
```

```python
from sklearn.preprocessing import OneHotEncoder

cat_cols = [
    "job_title", "employment_type",
    "company_location", "company_size",
    "education_required", "industry"
]

encoder = OneHotEncoder(handle_unknown="ignore", sparse_output=False)
cat_encoded = pd.DataFrame(
    encoder.fit_transform(data[cat_cols])
)
```

```python

num_cols = [
    "years_experience",
    "remote_ratio",
    "job_description_length",
    "benefits_score"
]

num_data = data[num_cols]
```

```python

X = pd.concat([skills_encoded, cat_encoded, num_data], axis=1)
y_salary = data["salary_usd"]
```

```python

def ats_score(row):
    return min(1, row["years_experience"] / 10)

data["ats_score"] = data.apply(ats_score, axis=1)
y_ats = data["ats_score"]
```

```python
data = data.sort_values("posting_date")

split1 = int(0.7 * len(data))
split2 = int(0.85 * len(data))

X_train, X_val, X_test = X[:split1], X[split1:split2], X[split2:]
y_train_ats, y_val_ats, y_test_ats = y_ats[:split1], y_ats[split1:split2], y_ats[split2:]
y_train_sal, y_val_sal, y_test_sal = y_salary[:split1], y_salary[split1:split2], y_salary[split2:]
```

```python
from sklearn.ensemble import RandomForestRegressor

ats_model = RandomForestRegressor(n_estimators=100)
# Convert all column names of X_train to string type
X_train.columns = X_train.columns.astype(str)
ats_model.fit(X_train, y_train_ats)
```

```python

from sklearn.ensemble import GradientBoostingRegressor

salary_model = GradientBoostingRegressor()
salary_model.fit(X_train, y_train_sal)
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np

# Convert all column names of X_test to string type
X_test.columns = X_test.columns.astype(str)

# ATS
ats_pred = ats_model.predict(X_test)
print("ATS RMSE:", np.sqrt(mean_squared_error(y_test_ats, ats_pred)))

# Salary
sal_pred = salary_model.predict(X_test)
print("Salary RMSE:", np.sqrt(mean_squared_error(y_test_sal, sal_pred)))
```
