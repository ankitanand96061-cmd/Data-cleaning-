```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plit
import seaborn as sns

data = pd.read_csv("ai_job_dataset.csv")
data
```

```python
data.head(2)
```

```python
data.shape
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.duplicated().sum()
```

```python

data['job_title'] = data['job_title'].str.lower()
data['company_name'] = data['company_name'].str.strip()
```

```python
data['posting_date'] = pd.to_datetime(data['posting_date'])
data['application_deadline'] = pd.to_datetime(data['application_deadline'])

data.dtypes
```

```python

sns.boxplot(x=data['salary_usd'])
plit.title("Salary Distribution")
plit.show()
```

```python

plit.figure()
plit.hist(data['salary_usd'], bins=20)
plit.title("Salary Distribution")
plit.xlabel("Salary USD")
plit.ylabel("Number of Jobs")
plit.show()
```

```python

plit.figure()
sns.barplot(x='experience_level', y='salary_usd', data=data)
plit.title("Experience Level vs Salary")
plit.show()
```

```python

plit.figure()
sns.barplot(x='experience_level', y='salary_usd', data=data)
plit.title("Experience Level vs Salary")
plit.show()
```

```python

plit.figure()
data['company_location'].value_counts().head(10).plot(kind='bar')
plit.title("Top Hiring Countries")
plit.xlabel("Country")
plit.ylabel("Number of Jobs")
plit.show()
```

```python

plit.figure()
sns.countplot(x='remote_ratio', data=data)
plit.title("Remote Work Distribution")
plit.show()
```

```python
data.to_csv("cleaned_jobs.csv", index=False)
data.head(3)
```
